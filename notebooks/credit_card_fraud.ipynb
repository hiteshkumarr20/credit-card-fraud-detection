```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/creditcard.csv")
```

```python
# Drop Time column
data = data.drop(['Time'], axis=1)
```

```python
# Separate features and target
X = data.drop('Class', axis=1)  # features
y = data['Class']               # target
```

```python
# Check shapes
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X['Amount'] = scaler.fit_transform(X[['Amount']])

# Check first 5 rows
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Fit SMOTE on training data
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Check new class distribution
print("Original y_train counts:\n", y_train.value_counts())
print("Resampled y_train counts:\n", y_train_res.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Initialize the model
lr = LogisticRegression(max_iter=1000, random_state=42)

# Train the model
lr.fit(X_train_res, y_train_res)

# Predict on test set
y_pred_lr = lr.predict(X_test)

# Evaluate
print("Logistic Regression Classification Report:\n")
print(classification_report(y_test, y_pred_lr))

# ROC-AUC score
y_pred_prob_lr = lr.predict_proba(X_test)[:,1]  # probabilities for class 1
roc_auc_lr = roc_auc_score(y_test, y_pred_prob_lr)
print("ROC-AUC Score:", roc_auc_lr)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize model
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# Train
rf.fit(X_train_res, y_train_res)

# Predict
y_pred_rf = rf.predict(X_test)

# Evaluate
print("Random Forest Classification Report:\n")
print(classification_report(y_test, y_pred_rf))

# ROC-AUC score
y_pred_prob_rf = rf.predict_proba(X_test)[:,1]
roc_auc_rf = roc_auc_score(y_test, y_pred_prob_rf)
print("ROC-AUC Score:", roc_auc_rf)
```

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    random_state=42,
    eval_metric='logloss'
)

xgb_model.fit(X_train_res, y_train_res)

y_pred_xgb = xgb_model.predict(X_test)

print("XGBoost Classification Report:\n")
print(classification_report(y_test, y_pred_xgb))

y_pred_prob_xgb = xgb_model.predict_proba(X_test)[:,1]
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_prob_xgb))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get feature importances from Random Forest
importances = rf.feature_importances_
features = X.columns

# Create a DataFrame for plotting
feat_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
feat_importance = feat_importance.sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='Importance', y='Feature', data=feat_importance)
plt.title("Random Forest Feature Importance")
plt.show()
```
